```python
def trimming(x):
    
    entries_to_remove = ('created_at', 'updated_at', 'created_by', 'updated_by', 'name_slug', 'idx')
#     for i in range(len(x)):
#         each = x[i]
    for e in range(len(x)):
        for k in entries_to_remove:
            x[e].pop(k, None)
    return x
```
