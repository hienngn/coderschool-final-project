```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
import html
import time
from urllib.parse import unquote

# optional
import google.datalab
from jupyter_datatables import init_datatables_mode
init_datatables_mode()
```

```python
homepage_r = requests.get('https://tiki.vn/')
homepage_soup = BeautifulSoup(homepage_r.text, 'html.parser')
homepage_menu_item = homepage_soup.body.find_all('li', {'class':'MenuItem-sc-181aa19-0 JgVAD'})
category = []

#Get Level 1 Category from Homepage
for i in homepage_menu_item:
    try:
        cat = {'cat_lv1':'', 'cat_lv1_url':'', 'cat_lv2':'', 'cat_lv2_url':'', 'cat_lv3':'', 'cat_lv3_url':''}
        cat['cat_lv1'] = i.text
        cat['cat_lv1_url'] = i.find('a', {'class': 'MenuItem__MenuLink-sc-181aa19-1 fKvTQu'})['href']
        cat['cat_lv2'] = ''
        cat['cat_lv2_url'] = ''
        cat['cat_lv3'] = ''
        cat['cat_lv3_url'] = ''
        category.append(cat)
    except:
        print('Error')
        
category = category
print(category)
```

```python
#Get Level 2 Category by loop through level 1 category url
for i in category:
    category_r = requests.get(i['cat_lv1_url'])
    category_soup = BeautifulSoup(category_r.text, 'html.parser')
    parent_cat = category_soup.find_all('div', {'class': 'list-group-item is-child'})
    parent_cat_name = i['cat_lv1']
    parent_cat_url = i['cat_lv1_url']
    for j in parent_cat:
        temp = []
        cat_lv2 = ' '.join(j.text.split()[:-1])
        cat_lv2_url = 'https://tiki.vn' + j.find('a',{'class': 'list-group-item'})['href']
        cat = {'cat_lv1':'', 'cat_lv1_url':'', 'cat_lv2':'', 'cat_lv2_url':'', 'cat_lv3':'', 'cat_lv3_url':''}
        try:
            cat['cat_lv1'] = parent_cat_name
            cat['cat_lv1_url'] = parent_cat_url
            cat['cat_lv2'] = cat_lv2
            cat['cat_lv2_url'] = cat_lv2_url
            cat['cat_lv3'] = ''
            cat['cat_lv3_url'] = ''
            temp.append(cat)
        except:
            print('Error')
        category = category + temp

print(category)
```

```python
#Get Level 3 Category by loop through level 2 categories url
temp_cat = [n for n in category if n['cat_lv2_url'] != '']
for i in temp_cat:
    category_r = requests.get(i['cat_lv2_url'])
    category_soup = BeautifulSoup(category_r.text, 'html.parser')
    parent_cat = category_soup.find_all('div', {'class': 'list-group-item is-child'})
    lv1_cat = i['cat_lv1']
    lv1_cat_name = i['cat_lv1_url']
    parent_cat_name = i['cat_lv2']
    parent_cat_url = i['cat_lv2_url']
    for j in parent_cat:
        temp = []
        cat_lv3 = ' '.join(j.text.split()[:-1])
        cat_lv3_url = 'https://tiki.vn' + j.find('a',{'class': 'list-group-item'})['href']
        cat = {'cat_lv1':'', 'cat_lv1_url':'', 'cat_lv2':'', 'cat_lv2_url':'', 'cat_lv3':'', 'cat_lv3_url':''}
        try:
            cat['cat_lv1'] = lv1_cat
            cat['cat_lv1_url'] = lv1_cat_name
            cat['cat_lv2'] = parent_cat_name
            cat['cat_lv2_url'] = parent_cat_url
            cat['cat_lv3'] = cat_lv3
            cat['cat_lv3_url'] = cat_lv3_url
            temp.append(cat)
        except:
            pass
        category = category + temp

tiki_category = pd.DataFrame(category)
tiki_category.to_pickle("./tiki_category.pkl")
```

```python
def scrap_page(cat_url, page_nbr):
    url = cat_url + '&page=' + str(page_nbr)
    category_r = requests.get(url)
    category_soup = BeautifulSoup(category_r.text, 'html.parser')
    sample_product = category_soup.body.find('div', {'class':'product-item'})
    all_product = category_soup.body.find_all('div', {'class':'product-item'})
    return all_product

def get_review(i):
    review = i.find('p', {'class':'review'}).text
    review = review.replace('(', '').split(' ')[0]
    review = int(review.replace('Chưa', '0'))
    return review

def get_link(i):
    link = i.find('a', {'class': ''})['href']
    if link[0:2] == '//':
        link_index = link.index('redirect=https%3A%2F%2F') + len('redirect=https%3A%2F%2F')
        link = unquote(link[link_index:])
    else:
        link = 'tiki.vn' + link
    return link

def get_img(i):
    img = i.find('img')['src']
    return img


def fetch_product(all_product, page_nbr):
    for i in all_product:
        d = {'site':'', 'sku':'', 'product_name':'', 'product_category':'', 'page_nbr': '','product_brand':'', 'price':'', 'review':'','url':'', 'img':''}
        try:
            d['site'] = 'tiki.vn' 
            d['sku'] = i['product-sku']
            d['product_name'] = i['data-title']
            d['product_category'] = i['data-category']
            d['page_nbr'] = page_nbr
            d['product_brand'] = i['data-brand']
            d['price'] = i['data-price']
            d['review'] = get_review(i)
            d['url'] = get_link(i)
            d['img'] = get_img(i)
            data.append(d)
        except:
            print('Error')
```

```python
data = []
tiki_product = []
for i in category_nt['cat_lv3_url']:
    for j in range(1,1000):
        all_product = scrap_page(i, j)
        if all_product == []:
            print('Break at: ' + str(i) + '&page=' + str(j))
            break
        fetch_product(all_product, j)
        print('Done: '  + str(i) + '&page=' + str(j))
        time.sleep(10)
tiki_product = pd.DataFrame(data)
tiki_product.to_pickle('./tiki_product.pkl')
```
