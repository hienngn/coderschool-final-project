## **Cleaning function**

```python
def clean_function_inci(x):
    element = x
    print(element)
    if element is np.nan:
        print('null value at: ', idx)
        return np.nan
    list_func = element.split(',')
    if len(list_func) > 1:
        functions = ''
        for index, func in enumerate(list_func):

            name = func.split(':')[0]
            name = name.replace("'", '').replace("{", "").strip().upper()
            if index == (len(list_func)-1):
                functions += name
            else:
                functions = functions + name + ', '
        return functions
    else:
        name = list_func[0].split(':')[0]
        functions = name.replace("'", '').replace("{", "").strip().upper()
        return functions
```
