```python
def get_url(url):
    """Get parsed HTML from url
      Input: url to the webpage
      Output: Parsed HTML text of the webpage
    """
    # Send GET request
    r = requests.get(url)

    # Parse HTML text
    soup = BeautifulSoup(r.text, "lxml")

    return soup
```

## **All in one**

```python
def scraping_ingredient(url):
    d = {'Ingredient_name': '',
         'Link': '',
        'Other_names': '',
        'Rating': '',
        'Irritancy': '',
        'Comedogenicity': '',
        'Functions': '',
        'Quick_facts': '',
        'Details': ''
        }
    d['Link'] = url
    soup = get_ingredient_url(url)
    
    get_name = soup.find('h1', {'class': 'klavikab lilac'})
    d['Ingredient_name'] = get_name.get_text().strip() if get_name else "None"

    get_rating = soup.find('div', {'class':'subtitle grey1'})
    d['Rating'] = get_rating.get_text().strip() if get_rating else "No rating"
    
    extra_divs = soup.find_all('div', {'class': 'itemprop'})
    for item in extra_divs:
        title = item.find('span', {'class':'label klavikab grey1'}).get_text().strip()
        if title == 'Irritancy:':
            d['Irritancy'] = item.find('span', {'class': 'value'}).get_text().strip() 
        elif title == 'Comedogenicity:':
            d['Comedogenicity'] = item.find('span', {'class': 'value'}).get_text().strip()
        elif title == 'Also-called-like-this:':
            d['Other_names'] = item.find('span', {'class': 'value'}).get_text().strip() 
        else:
            d['Functions'] = {url.get_text():url['href'] for url in item.find_all('a')}
    
    quick_facts = []
    for content in soup.select("[id^='showmore-section-quickfacts'] ul"):
        get_fact = content.text + '.'
        quick_facts.append(get_fact.replace('\n','').strip())
        if not quick_facts:
            d['Quick_facts'] = 'No information'
        else:         
            d['Quick_facts'] = quick_facts
    
    details = []
    for content in soup.select("[id^='showmore-section-details'] p"):
        get_detail = content.text
        details.append(get_detail.replace('\n','').strip())
        if not details:
            d['Details'] = 'No information'
        else:         
            d['Details'] = details
    
    return d
    
    
```
