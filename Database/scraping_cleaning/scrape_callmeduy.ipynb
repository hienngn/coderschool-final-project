```python
def name_trimming(x):
    
    entries_to_remove = ('created_at', 'updated_at', 'created_by', 'updated_by', 'name_slug', 'idx')
#     for i in range(len(x)):
#         each = x[i]
    for e in range(len(x)):
        for k in entries_to_remove:
            x[e].pop(k, None)
    return x

def create_category(x):
    category_id = []
    
    if not x:
        category_id = 'Pending'
    for each in x:
        category_id.append(each['id'])
        
    return category_id

def create_function(x):
    function = []
    
    if not x:
        function = 'Pending'
    for each in x:
        function.append(each['title'])  
        
    return function

def remove_html_tag(x):
    res = re.sub("<.*?>", " ", str(x))
    return res
```

```python
ingredients2 = ingredients.copy()
```

```python
ingredients2['names'] = ingredients2['names'].apply(name_trimming)
ingredients2['category_id'] =ingredients2['tags'].apply(create_category)
ingredients2['functions'] =ingredients2['tags'].apply(create_function)
ingredients2['description'] = ingredients2['description'].apply(remove_html_tag)
ingredients2.drop(columns=['product', 'tags'], inplace=True)
ingredients2.reset_index(drop=True, inplace=True)
ingredients2.tail()
```

```python
ingredient_df = pd.DataFrame(columns=['chemical_id', 'id', 'name', 'description', 'rating', 'category_id', 'functions'])
for each in tqdm(names):
    for i in range(len(each)):
        ing = each[i]
#         print(ing)
        res_dict = {'chemical_id': ing['chemical'], 
                    'id': ing['id'], 
                    'name': ing['name'], 
                    'description': ingredients2[ingredients2['id'] == ing['chemical']]['description'].values, 
                    'rating': ingredients2[ingredients2['id'] == ing['chemical']]['rating'].values, 
                    'category_id': ingredients2[ingredients2['id'] == ing['chemical']]['category_id'].values, 
                    'functions':ingredients2[ingredients2['id'] == ing['chemical']]['functions'].values
                    }
        ingredient = pd.DataFrame(res_dict)
        ingredient_df = ingredient_df.append(ingredient)
print(ingredient_df.shape)
```
