## **Scrape: Skin**

```python
regex = re.compile('product-score-img.*')
```

```python
product_list = []
for link in list_hair:
    for page_num in tqdm(range(1, 200)):
        soup = selenium_get_url(link + f'?page={page_num}')
        products = soup.find_all('div', {'class': 'product-tile'})
        
        if len(products) == 0:
            print('Break at: ' + str(link) + f'?page={page_num}') #&per_page=36
            break
        
        else:
            for i in range(len(products)):
                d = {'cat_name': '',
                    'subcat_name': '',
                    'product_url': '',
                    'product_brand': '',
                    'product_name': '',
                    'product_score': '',
                    'product_img': ''}
                try: 
                    d['cat_name'] = df_category[df_category['link'] == link]['cat_name'].to_list()[0]
                    d['subcat_name'] = df_category[df_category['link'] == link]['subcat_name'].to_list()[0]
                    d['product_img'] = products[i].find('img', {'class': 'product-image'})['src'] #image URL
                    d['product_url'] = products[i].find('a')['href'] # product URL
                    d['product_score'] = products[i].find('img', {'class': regex})['src'].split('score')[-1].split('-')[1] # product score
                    d['product_brand'] = products[i].find('p', {'class': 'product-company'}).get_text() # brand
                    d['product_name'] = products[i].find('p', {'class': 'product-name'}).get_text().strip() # product name
                    product_list.append(d)
                
                except:
                    print('error at: ', products[i])
        
        df_products = pd.DataFrame(product_list)
        df_products.to_csv('ewg_hair.csv', index=False)
        
        time.sleep(1)
```
